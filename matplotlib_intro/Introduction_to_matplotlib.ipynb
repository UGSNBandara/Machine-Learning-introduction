### Lets plot using OO methos 

Which one we should use, OO method or pyplot method ? 

    When ploting something quickly, ok to use the pyplot method
    when ploting something advanced, use the OO method 

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind="scatter", x = "age", y ="chol", c= "target", figsize = (10,6))
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax)
ax.set(title="Age vs Cholostorol level Variation")

### we can limit the x axis & also possible to y axis 
ax.set_xlim([60,75])

ax.set_xlabel("Age variation")


## when use the OO method should gives the ax
```

```python
over_50.target.values
```

```python
fig, ax = plt.subplots(figsize = (10,6))

##important to assign graph to a variable to some modifications
scatter_ = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]);

ax.set(title="Age and Chol level variation among over 50 patients",
      xlabel="Age ",
      ylabel="Chol level");

#ax.set_xlim([55,70])

ax.legend(*scatter_.legend_elements(), title="Target" )

#lets add a horizontal line which is goes through the mean value of cholesterol level
ax.axhline(y=over_50.chol.mean(),
          linestyle = "-.")

## "-", "--", "-.", ""
```

```python
over_50.chol.mean()
```

```python
over_50
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10,11))

age_chol = ax0.scatter(over_50["age"], over_50["chol"], c = over_50["target"])
ax0.legend(*age_chol.legend_elements(), title="Target")
ax0.set(title="Age vs Cholesterol level",
        ylabel="Cholesterol level")

##horizontal line through the mean value.
ax0.axhline(y = over_50["chol"].mean(), linestyle = "--")


## define axis and soo
age_thalach = ax1.scatter(over_50["age"], over_50["thalach"], c = over_50["target"])
ax1.legend(*age_thalach.legend_elements(), title = "Target")
ax1.set(title = "Age vs Max heart rate", 
       xlabel="Age", ylabel="Max heart rate")

## horizontal line through the mean value 
ax1.axhline(y = over_50["thalach"].mean(), linestyle = "--")

##  add a title for the figure
fig.suptitle("Heart Disease Analysis", fontsize = 15, fontweight = "bold")
### it is suptitile, not sub *****

#saving the figure as .....png
fig.savefig("Age_chol_thalach_variation_with_target.png")
```
