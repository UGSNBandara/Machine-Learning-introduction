## Matplotlib work flow 

```python
#0. import matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

#1 prepare data
x = [1,2,3,4,5]
y = [1,2,3,4,5]

#2 Setyp plot || OO 
fig, ax = plt.subplots(figsize=(5,6)) #width and high 

#3 Plot data
ax.plot(x,y)

#4 Costomize plot
ax.set(title = "test plot", 
       xlabel = "X axis", 
       ylabel =  "Y axis")

#5 save & show
fig.savefig("sample.png")
```

Exampple  :

## Marking figures with numpy arrays

We want:
*line plot
*scatter plot
*bar plot
*histrogram
*subplot

```python
x = np.linspace(0,10,50)
fig, ax = plt.subplots(figsize = (5,5))

ax.plot(x, x**2)
```

```python
#usigng same data lets create
fig, ax = plt.subplots(figsize = (5,4))

ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, np.sin(x))
```

```python
#lets plot a dictionary as bar plot
sales = {"BMW" : 6, "GTR" : 8, "Benz" : 3, "Tesla" : 6,}
fig, ax = plt.subplots()
ax.bar(sales.keys(),sales.values())

ax.set(title="Car sales Bar Chart",
      xlabel="Car Brands", ylabel="Number of cars sold")
```

```python
fig, ax = plt.subplots()
ax.barh(list(sales.keys()),list(sales.values()))

ax.set(title="Car sales Horizontal Bar Chart",
      ylabel="Car Brands", xlabel="Number of cars sold")
```

```python
#make some data for histogram 
x = np.random.randn(1000)
# gives the standard normal distribution
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(x)
```

### sub plot options 1 : 

```python
fig, ((ax0, ax1),(ax2, ax3), (ax4, ax5)) = plt.subplots(figsize = (10,10), nrows = 3, ncols=2)

ax1.plot([1,2,3,4,5,6])
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(["Choco", "Vanilla", "Stuabary"], [10,25,10])
ax4.barh(["Choco", "Vanilla", "Stuabary"], [10,25,10])
ax5.scatter(np.linspace(0,100,50), np.square(np.linspace(1,100,50)))
ax0.scatter(np.linspace(0,100,50), np.sin(np.linspace(1,100,50)))
```

### Sub plot option 2 :

```python
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                    figsize = (10,10))

## to plot in different axes; can access  axes, using
## ax[0,0]  -- ax[0,1] -- ax[1,0]  -- ax[1,1]

ax[0, 0].plot([1,2,3,4,5,6],[0.5,1,1.5,2,2.5,3]);
ax[0, 1].bar(["Choco", "Vanilla", "Stuabary"], [10,25,10])
ax[1, 0].barh(["Choco", "Vanilla", "Stuabary"], [10,25,10])
ax[1, 1].scatter(np.linspace(0,100,50), np.sin(np.linspace(1,100,50)))
```
